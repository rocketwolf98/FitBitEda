#### III. BMI Influence in Medal Success

General coaching staff from various professional athelete management firms seek to understand correlation of body composition and performance outcome in a given field of sports. Your team was asked to provide data on the average BMI class of gold medalists per sports discipline.

##### BMI formula

<img src="images/bmi_formula.png" alt="BMI Formula" width="500" />

##### Standard BMI classification

<img src="images/bmi_class.png" alt="BMI Classification" width="500" />

```python
# Start wrangling here
olympics_bios.rename(columns={'NOC': 'noc'}, inplace=True)
olympics_bios.head()
bios_results_merged = pd.merge(olympics_results_gold, olympics_bios, on ='noc', how='left')
bios_results_merged_gold = bios_results_merged[bios_results_merged['medal']=='Gold']
bios_results_merged_gold
```

```python
bios_results_merged_gold.info()
```

```python
bios_na_drop = bios_results_merged_gold
```

```python
bios_na_drop['height_cm'].isna().sum()
```

```python
bios_na_drop['weight_kg'].isna().sum()
```

```python
bios_na_drop = bios_na_drop.dropna(subset=['weight_kg'])
```

```python
bios_na_drop.info()
```

```python
bios_cleaned = bios_na_drop[['discipline', 'medal', 'height_cm', 'weight_kg']]
bios_cleaned
```

```python
bios_cleaned['bmi'] = bios_cleaned['weight_kg'] / (bios_cleaned['height_cm'] * 0.01) ** 2
```

```python
bios_cleaned
```

```python
def cat_bmi(row):
  if row['bmi'] < 18.5:
    return 'Underweight'
  elif 18.5 <= row['bmi'] < 25.0:
    return 'Normal'
  elif 25.0 <= row['bmi'] < 30.0:
    return 'Overweight'
  else:
    return 'Obese'

bios_cleaned['weight_category'] = bios_cleaned.apply(cat_bmi, axis=1)
bios_cleaned.head()
```

```python
bios_cleaned.info()
```

```python
bios_cleaned.groupby(['discipline'])['bmi'].mean()
```
